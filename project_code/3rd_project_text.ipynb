```python
from nltk import Text
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from project.crawling import crawling
from project.cleaning_okt import cleaning_okt
from project.visualization import visualization
# from project.wordcloud import wordcloud
import pandas as pd
import os

# í¬ë¡¤ë§í•  URL
url = 'https://www.youtube.com/watch?v=kW_z-NMuZIU'
# í¬ë¡¤ë§ ì‹¤í–‰
comment_list = crawling(url)

# í¬ë¡¤ë§ ê²°ê³¼ ì¶œë ¥
print('í¬ë¡¤ë§ ê²°ê³¼ : ', comment_list)

# í¬ë¡¤ë§í•œ ëŒ“ê¸€ íŒŒì¼ ì €ìž¥ 
os.makedirs("project", exist_ok=True)
file_name = 'project/comments.txt'

with open(file_name, 'wt', encoding='utf-8') as file:
    file.write(".".join(comment_list))
comments = str(comment_list).replace("\n"," ")


# ë°ì´í„° ì •ì œ
### Oktë¡œ í† í°í™”
result1 = cleaning_okt(comments)
### WordPieceë¡œ í† í°í™”
# result2 = cleaning_wp(comments)


# ë¹ˆë„ìˆ˜ ê·¸ëž˜í”„ ì‹œê°í™”
visualization(result1)

text1 = Text(result1, name='í˜¸ì£¼ì „ - Okt')
# text2 = Text(result2, name='í˜¸ì£¼ì „ - WP')


# ë¹ˆë„ìˆ˜ ê´€ë ¨ ë¶„ì„
fd = text1.vocab()
print('ì´ í† í° ê°œìˆ˜ : ', fd.N())
print('ê³ ìœ  í† í° ê°œìˆ˜ : ', fd.B())
print('ê°€ìž¥ ë§Žì´ ë‚˜ì˜¨ í† í° : ', fd.max())
print('\'ì†í¥ë¯¼\'ì˜ ë¹ˆë„ìˆ˜ : ', fd.get('ì†í¥ë¯¼'))
print('\'í™©í¬ì°¬\'ì˜ ë¹ˆë„ìˆ˜ : ', fd.get('í™©í¬ì°¬'))
print(f'ê°€ìž¥ ë§Žì´ ë‚˜ì˜¨ í† í°ì˜ ë¹ˆë„ìˆ˜ : {fd.get(fd.max())}')
print(f'ê°€ìž¥ ë§Žì´ ë‚˜ì˜¨ í† í°ì˜ ì´ í† í° ìˆ˜ ëŒ€ë¹„ ë¹„ìœ¨ : {fd.freq(fd.max()):.2%}')


# WordCloud ìƒì„±
# wordcloud(fd)


# train / test ë°ì´í„°ì…‹ ë‚˜ëˆ„ê¸°
train_set, test_set = train_test_split(comment_list, test_size=0.2)

# Count
# cv = CountVectorizer(
#     min_df = 3,
#     tokenizer = cleaning_okt,
#     token_pattern = None
# )
# cv.fit(train_set + test_set)  # DTMì„ ì–´ë–»ê²Œ ë§Œë“¤ì§€ train setì„ ì´ìš©í•´ í•™ìŠµ.  ==> vocab ìƒì„±
# train_cv = cv.transform(train_set)
# test_cv = cv.transform(test_set) # ë³€í™˜ìž‘ì—…  ==> ë¬¸ìž¥ë³„ ë¹ˆë„ìˆ˜ë¥¼ ê°€ì§€ê³  dtmì„ ìƒì„±.

# train_df_dtm = pd.DataFrame(train_cv.toarray(), columns=cv.get_feature_names_out())
# test_df_dtm = pd.DataFrame(test_cv.toarray(), columns=cv.get_feature_names_out())

# TfIdf
tfidf = TfidfVectorizer(
    max_features=150,
    min_df = 10,
    tokenizer = cleaning_okt,
    token_pattern = None
)
tfidf.fit(train_set + test_set)
train_tf = tfidf.transform(train_set)
test_tf = tfidf.transform(test_set)

train_df_tf = pd.DataFrame(train_tf.toarray(), columns=tfidf.get_feature_names_out())
test_df_tf = pd.DataFrame(test_tf.toarray(), columns=tfidf.get_feature_names_out())
test_df_tf
```
